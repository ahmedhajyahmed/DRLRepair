```python
import numpy as np

arr = np.load("internals.npz", allow_pickle=True)
ob_bef = arr['ob']
weights_bef = [arr['0'], arr['2'], arr['4']]
biases_bef = [arr['1'], arr['3'], arr['5']]
```

```python
arr = np.load("experiments/internals_fm_5_m_5.npz", allow_pickle=True)
ob_aft = arr['ob']
weights_aft = [arr['0'], arr['2'], arr['4']]
biases_aft = [arr['1'], arr['3'], arr['5']]
```

```python
b = weights_aft[0]
a = weights_bef[0][:len(b)]


print(a.shape)
print(b.shape)

c = b - a
print()
```

```python
mins_c = np.min(c, axis=(1, 2))
maxs_c = np.max(c, axis=(1, 2))
stds_c = np.std(c, axis=(1, 2))
means_c = np.mean(c, axis=(1, 2))
medians_c = np.mean(c, axis=(1, 2))
```

```python
import matplotlib.pyplot as plt

plt.title("mins")
plt.plot(mins_c)
plt.show()
plt.title("maxs")
plt.plot(maxs_c, color='yellow')
plt.show()
plt.title("stds")
plt.plot(stds_c, color='green', linestyle='-')
plt.show()
plt.title("means")
plt.plot(means_c, color='red', linestyle='-')
plt.show()
plt.title("medians")
plt.plot(medians_c, color='blue', linestyle='-')
plt.show()
```
